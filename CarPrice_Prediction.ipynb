Assing the columns of x to input labels i.e features and y to output labels i.e target 

```python
x_columns=['Year','Selling_Price','Driven_kms','Owner']
y_columns=['Selling_Price']

x = df[x_columns]
y = df[y_columns]
```

```python

x = x.to_numpy().reshape(-1, len(x_columns))
y = y.to_numpy().reshape(-1, len(y_columns))
```

```python
#x_encoded = pd.get_dummies(x, columns=['Fuel_Type', 'Transmission'], drop_first=True)
```

splitting the data into 80% training and 20% into  test data for prediction of model

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,train_size=0.8)
```

Linear regression is used as it of continous value type

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
print(x_train,x_test,y_train,y_test)
```

fitting the model

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

predicting the model for test data

```python
y_pred = lr.predict(x_test)
```

These metrics help you assess how well the model generalizes to unseen data and whether it provides accurate predictions for car prices. 
Higher R2 values indicate better model performance, as they suggest that the model explains more of the variability in car prices.
MSE measures the average squared difference between the predicted car values and the actual car values.
MAE measures the average absolute difference between the predicted values and the actual values.
MAE measures the average absolute difference between the predicted values and the actual values.


```python
import sklearn.metrics as sm
print("Mean absolute error =", (sm.mean_absolute_error(y_test, y_pred)))
print("Mean squared error =", (sm.mean_squared_error(y_test, y_pred)))
print("R^2 score =", (sm.r2_score(y_test, y_pred)))
```
